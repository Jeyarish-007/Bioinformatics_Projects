# TITLE: Mutation Visualization In Aligned DNA Sequences (SNPs)

```python
import matplotlib.pyplot as plt
from Bio.Seq import Seq

# Define sequences
seq1 = Seq("ATGCGTACGTA")
seq2 = Seq("ATGCGTACCTA")

# Find mutations
mutations = [(i, seq1[i], seq2[i]) for i in range(len(seq1)) if seq1[i] != seq2[i]]

# Visualization 1: Bar Chart of Mutations
positions = [m[0] for m in mutations]
original_bases = [m[1] for m in mutations]
mutated_bases = [m[2] for m in mutations]

plt.figure(figsize=(8, 4))
plt.bar(positions, [1] * len(positions), color='red', alpha=0.6)
plt.xticks(positions, [f"{o}->{m}" for o, m in zip(original_bases, mutated_bases)])
plt.xlabel("Position")
plt.ylabel("Mutation")
plt.title("Mutation Positions in DNA Sequence")
plt.show()

# Visualization 2: Sequence Alignment with Mutations Highlighted
seq1_str = str(seq1)
seq2_str = str(seq2)

aligned_seq1 = ""
aligned_seq2 = ""
highlight = ""

for i in range(len(seq1_str)):
    if seq1_str[i] == seq2_str[i]:
        aligned_seq1 += seq1_str[i]
        aligned_seq2 += seq2_str[i]
        highlight += " "
    else:
        aligned_seq1 += f"\033[91m{seq1_str[i]}\033[0m"
        aligned_seq2 += f"\033[91m{seq2_str[i]}\033[0m"
        highlight += "^"

print("\nSequence Alignment:")
print("Seq1:", aligned_seq1)
print("Seq2:", aligned_seq2)
print("      ", highlight)
```
